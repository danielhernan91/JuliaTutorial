# PLotly no vale

```julia
Pkg.add("Plotly")
```

```julia
using Plotly
```

```julia
trace1 = [
  "x" => [1, 2, 3, 4],
  "y" => [10, 15, 13, 17],
  "type" => "scatter"
]
trace2 = [
  "x" => [1, 2, 3, 4],
  "y" => [16, 5, 11, 9],
  "type" => "scatter"
]
data = [trace1, trace2]
response = Plotly.plot(data, ["filename" => "basic-line", "fileopt" => "overwrite"])
plot_url = response["url"]
```

```julia
data = [
  [
    "z" => [1 20 30; 20 1 60; 30 60 1],
    "type" => "heatmap"
  ]
]
response = Plotly.plot(data, ["filename" => "basic-heatmap", "fileopt" => "overwrite"])
plot_url = response["url"]
```
